```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#loading data
#dataset is from kaggle - bank notes
df= pd.read_csv('bank_note_data.csv')

#data is 4 dimensional - 4 features
#target is to convert 4d data into 2d data
features=['Image.Var','Image.Skew','Image.Curt','Entropy']
target=['Class']

#scaling data
data_norm=df.copy()
sc=StandardScaler()
data_norm[features]=sc.fit(df[features]).transform(df[features])

#plot t-SNE
tsne=TSNE(n_components=2, learning_rate=500)
x_tsne=tsne.fit_transform(data_norm[features])
y_tsne=data_norm[target]

#visualization
plt.figure(figsize=(16,11))
plt.scatter(x_tsne[np.where(y_tsne==0),0],x_tsne[np.where(y_tsne==0),1],color='g',marker='*')
plt.scatter(x_tsne[np.where(y_tsne==1),0],x_tsne[np.where(y_tsne==1),1],color='r',marker='.')
plt.title('t-SNE')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```
